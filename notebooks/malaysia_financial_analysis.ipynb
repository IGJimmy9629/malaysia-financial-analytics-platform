```python
# ============================================================
# Malaysia Financial and Economic Indicators Analysis
# Using Apache Hive Output + Python in VS Code Jupyter Notebook
# ============================================================

# Import the required Python libraries.
# pandas is used for data cleaning and analysis.
# matplotlib is used to create visualisations.
# os is used to create folders for processed data and charts.

import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Create folders to store processed datasets and visual outputs.
# If the folders already exist, Python will not create duplicates.

os.makedirs("../data/processed", exist_ok=True)
os.makedirs("../visuals", exist_ok=True)
```

```python
# Load the dataset exported from Apache Hive.
# This file contains the financial and economic indicators selected using Hive.

df = pd.read_csv(r"C:\Users\user\Documents\malaysia-financial-hive-project\data\processed\selected_indicators.csv")

# Display the dataset size and first few rows to understand the structure.

print("Dataset size:", df.shape)
display(df.head())
```

```python
# The exported Hive file has column names such as:
# selected_indicators.country_name
# selected_indicators.indicator_name
#
# This step removes the "selected_indicators." prefix so the column names are easier to use.

df.columns = df.columns.str.replace("selected_indicators.", "", regex=False)

# Display updated column names and preview the data.

print("Updated column names:")
print(df.columns.tolist())

display(df.head())
```

```python
# Identify all year columns.
# In the Hive output, year columns are named like y1960, y1961, y1962 and so on.

year_cols = [col for col in df.columns if col.startswith("y")]

print("Number of year columns:", len(year_cols))
print(year_cols[:5], "...", year_cols[-5:])
```

```python
# Convert the dataset from wide format to long format.
#
# Wide format:
# one row per indicator, many year columns.
#
# Long format:
# one row per indicator per year.
#
# Long format is easier for analysis and visualisation.

df_long = df.melt(
    id_vars=["country_name", "country_code", "indicator_name", "indicator_code"],
    value_vars=year_cols,
    var_name="year",
    value_name="value"
)

# Remove the letter "y" from the year column.
# Example: y2024 becomes 2024.

df_long["year"] = df_long["year"].str.replace("y", "").astype(int)

# Convert values into numeric format.
# Any non-numeric value will become NaN.

df_long["value"] = pd.to_numeric(df_long["value"], errors="coerce")

display(df_long.head())
```

```python
# Keep only recent years from 2000 to 2024.
# This period is selected because it gives a more current view of Malaysia's financial and economic trend.

df_clean = df_long[
    (df_long["year"] >= 2000) &
    (df_long["year"] <= 2024)
]

# Remove rows where value is missing.
# Missing values are removed so the visualisation and summary statistics are cleaner.

df_clean = df_clean.dropna(subset=["value"])

# Save the cleaned dataset for GitHub submission.

df_clean.to_csv("../data/processed/malaysia_financial_indicators_clean.csv", index=False)

print("Cleaned dataset size:", df_clean.shape)
display(df_clean.head())
```

```python
# Create summary statistics for each indicator.
# This gives a quick view of minimum, maximum, average and spread of the values.

summary = df_clean.groupby("indicator_name")["value"].describe()

display(summary)

# Save summary statistics.

summary.to_csv("../data/processed/summary_statistics.csv")
```

```python
# Create a pivot table.
# This format makes it easier to compare different indicators by year.

pivot_df = df_clean.pivot_table(
    index="year",
    columns="indicator_name",
    values="value"
)

display(pivot_df.tail(10))

# Save pivot table.

pivot_df.to_csv("../data/processed/pivot_financial_indicators.csv")
```

```python
# Get the latest available value for each indicator.
# This helps to identify the most recent condition of each indicator.

latest_values = (
    df_clean.sort_values("year")
    .groupby("indicator_name")
    .tail(1)
    .sort_values("indicator_name")
)

display(latest_values[["indicator_name", "year", "value"]])

# Save latest values.

latest_values.to_csv("../data/processed/latest_indicator_values.csv", index=False)
```

```python
# Create a reusable function for line charts.
# This avoids repeating the same plotting code many times.

def plot_indicator(indicator_name, filename):
    temp = df_clean[df_clean["indicator_name"] == indicator_name].sort_values("year")

    plt.figure(figsize=(10, 5))
    plt.plot(temp["year"], temp["value"], marker="o")
    plt.title(indicator_name)
    plt.xlabel("Year")
    plt.ylabel("Value")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"../visuals/{filename}", dpi=300)
    plt.show()
```

```python
# Visualisation 1: GDP growth
# This chart shows Malaysia's annual GDP growth trend.

plot_indicator("GDP growth (annual %)", "gdp_growth.png")
```

Analysis and Insight:

The GDP growth chart shows the movement of Malaysia's economic growth over time.
A higher value means the economy expanded more strongly, while a lower or negative
value shows weaker economic performance.

From this chart, the important point is to observe years where GDP growth dropped
sharply and years where it recovered. This is useful for the banking industry because
economic growth affects borrowing demand, investment activity and repayment ability.

```python
# Visualisation 2: Inflation
# This chart shows the annual change in consumer prices.

plot_indicator("Inflation, consumer prices (annual %)", "inflation.png")
```

Analysis and Insight:

The inflation chart shows how consumer prices changed over the years.
Higher inflation means prices increased faster, while lower inflation means prices
were more stable.

For banks and financial institutions, inflation is important because it affects interest
rates, household spending, savings behaviour and loan affordability. If inflation rises
too quickly, customers may face higher living costs and this can affect their ability
to repay loans.

```python
# Visualisation 3: Unemployment
# This chart shows the unemployment rate in Malaysia.

plot_indicator(
    "Unemployment, total (% of total labor force) (modeled ILO estimate)",
    "unemployment.png"
)
```

Analysis and Insight:

The unemployment chart shows the percentage of the labour force that is unemployed.
A higher unemployment rate may indicate pressure in the labour market.

For the finance and banking industry, unemployment is an important risk indicator.
When unemployment increases, some households may have difficulty maintaining loan
repayments. This can increase credit risk for banks, especially in personal financing,
housing loans and vehicle financing.

```python
# Visualisation 4: Domestic credit to private sector by banks
# This chart shows the amount of bank credit provided to the private sector as a percentage of GDP.

plot_indicator(
    "Domestic credit to private sector by banks (% of GDP)",
    "domestic_credit.png"
)
```

Analysis and Insight:

This chart shows how much banks are supporting the private sector through credit.
A higher value means banks are providing more financing to households and businesses
relative to the size of the economy.

This indicator is directly related to the banking industry. Strong domestic credit can
support business growth and consumer activity. However, very high credit growth also
needs to be monitored because it may increase financial risk if lending quality is weak.

```python
# Visualisation 5: Bank non-performing loans
# This chart shows the percentage of bank loans that are classified as non-performing.

plot_indicator(
    "Bank nonperforming loans to total gross loans (%)",
    "nonperforming_loans.png"
)
```

Analysis and Insight:

The non-performing loans chart shows the quality of bank lending.
A lower ratio is generally better because it means fewer loans are in repayment problem.

This is one of the most important indicators for banks. If non-performing loans increase,
banks may need to strengthen credit assessment, collection and risk monitoring. This
indicator also helps show whether the banking sector is stable or facing pressure from
bad loans.

```python
# Visualisation 6: Comparison of key economic indicators
# This chart compares GDP growth, inflation and unemployment in one graph.

comparison_indicators = [
    "GDP growth (annual %)",
    "Inflation, consumer prices (annual %)",
    "Unemployment, total (% of total labor force) (modeled ILO estimate)"
]

plt.figure(figsize=(11, 6))

for indicator in comparison_indicators:
    temp = df_clean[df_clean["indicator_name"] == indicator].sort_values("year")
    plt.plot(temp["year"], temp["value"], marker="o", label=indicator)

plt.title("Malaysia Key Economic Indicators")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("../visuals/key_economic_indicators_comparison.png", dpi=300)
plt.show()
```

Analysis and Insight:

This comparison chart gives a broader view of Malaysia's economy by placing GDP growth,
inflation and unemployment together.

The chart is useful because these indicators are connected. For example, weak GDP growth
may be linked with higher unemployment, while inflation affects purchasing power and
borrowing cost. For banks, looking at these indicators together can support better
decision-making in lending, risk management and business planning.
